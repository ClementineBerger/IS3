```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import os

from IPython.display import display, Audio

import librosa
```

```python
parent_data_folder = os.path.join(
    os.environ['AUDIBLE_DATA'],
    "rendering",
    "datasets",
)

# parent_data_folder = os.environ['MY_DATA']

metadata_path = os.path.join(
    parent_data_folder,
    "impulsive_sound_rejection",
    "synthetic_stationary_background",
    "metadata.csv"
)
```

```python
# Load the metadata, opening dataframe

metadata = pd.read_csv(metadata_path)
display(metadata)
```

```python
idx = np.random.randint(0, len(metadata))
file_path = os.path.join(parent_data_folder, metadata.iloc[idx]['file_path'])

print(file_path)

signal, sr = librosa.load(file_path, sr=None)
print(sr)
```

```python
plt.figure(figsize=(10, 5))
t = np.arange(len(signal)) / sr
plt.plot(t, signal)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Signal')
plt.grid()
plt.show()

display(Audio(signal, rate=sr))
```
